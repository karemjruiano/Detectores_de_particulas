```python
import numpy as np

# Parámetros iniciales
E0 = 100.0  
deltaX = 0.1  
# Función que calcula dE/dx a partir de la energía E
def dEdx(E):
    return 0.01 * E**2 - 1.43 * E + 51.48
E = E0
X = 0.0
results = []
while E > 0:
    # Pérdida de energía
    dE = dEdx(E) * deltaX
    E -= dE
    results.append((X, dE))
    X += deltaX
results_np = np.array(results)
R = X if E <= 0 else None
results_np, R
```

```python

def dEdx_new(E):
    return 97.74 * np.exp(-0.89 * E) + 5.20

E = E0
X = 0.0
results_new = []

while E > 0:
    dE = dEdx_new(E) * deltaX
    E -= dE
    results_new.append((X, dE))
    X += deltaX
results_np_new = np.array(results_new)
R_new = X if E <= 0 else None
results_np_new, R_new
```
